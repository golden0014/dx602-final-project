### Problem 6 (50 points)

Build three different regression models using the entire data set.
Plot the actual target vs the predicted values for each in one chart.
Compute the L2 and L1 losses for each of them.
You may use any regression class provided provided by scikit-learn, and you may reuse one class as long as you change its parameters enough to see different results.

```python
# YOUR CODE HERE
X = bodyfat.drop('target', axis = 1)
y = bodyfat['target']
```

```python
model1 = sklearn.linear_model.LinearRegression()
model1.fit(X, y)
y_pred1 = model1.predict(X)
print(f"L1 loss for Linear Regression model is {sklearn.metrics.mean_absolute_error(y, y_pred1):.3f}")
print(f"L2 loss for Linear Regression model is {sklearn.metrics.mean_squared_error(y, y_pred1):.3f}")

plt.scatter(y_pred1, y)
plt.title('Actual vs. Predicted Body Fat %')
plt.xlabel('Predicted Body Fat (%)')
plt.ylabel('Actual Body Fat (%)')
plt.show()
```

```python
model2 = sklearn.linear_model.Ridge() # leave alpha as default, 1.0
model2.fit(X, y)
y_pred2 = model2.predict(X)
print(f"L1 loss for Ridge Regression model is {sklearn.metrics.mean_absolute_error(y, y_pred2):.3f}")
print(f"L2 loss for Ridge Regression model is {sklearn.metrics.mean_squared_error(y, y_pred2):.3f}")

plt.scatter(y_pred2, y)
plt.title('Actual vs. Predicted Body Fat %')
plt.xlabel('Predicted Body Fat (%)')
plt.ylabel('Actual Body Fat (%)')
plt.show()
```

```python
model3 = sklearn.linear_model.Lasso() # leave alpha as default, 1.0
model3.fit(X, y)
y_pred3 = model3.predict(X)
print(f"L1 loss for Lasso Regression model is {sklearn.metrics.mean_absolute_error(y, y_pred3):.3f}")
print(f"L2 loss for Lasso Regression model is {sklearn.metrics.mean_squared_error(y, y_pred3):.3f}")

plt.scatter(y_pred3, y)
plt.title('Actual vs. Predicted Body Fat %')
plt.xlabel('Predicted Body Fat (%)')
plt.ylabel('Actual Body Fat (%)')
plt.show()
```

### Problem 7 (30 points)

Use 5-fold cross-validation to repeat building the same three kinds of regression models. Compare the L2 losses predicted by cross-validation against the L2 losses training against the whole data set. (The difference is likely from overfitting in the latter.)

```python
model1 = sklearn.linear_model.LinearRegression()
print(f"L2 loss for Linear Regression model is {-sklearn.model_selection.cross_val_score(model1, X, y, cv = 5, scoring = 'neg_mean_squared_error').mean():.3f}")
```

```python
model2 = sklearn.linear_model.Ridge()
print(f"L2 loss for Ridge Regression model is {-sklearn.model_selection.cross_val_score(model2, X, y, cv = 5, scoring = 'neg_mean_squared_error').mean():.3f}")
```

```python
model3 = sklearn.linear_model.Lasso()
print(f"L2 loss for Lasso Regression model is {-sklearn.model_selection.cross_val_score(model3, X, y, cv = 5, scoring = 'neg_mean_squared_error').mean():.3f}")
```

The L2 loss for the models using cross validation are higher than the models produced from training on the entire data set as predicted.

### Problem 8 (25 points)

Build three different regression models as in problem 6, but preprocess the data so that each column has mean zero and standard deviation one first.
For full credit, use a scikit-learn pipeline for each model.
For each model, compare the L2 losses -- which of them performed differently from your results in problem 6?

(This process will be covered in week 13.)

```python
# YOUR CODE HERE

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
pipe1 = sklearn.pipeline.Pipeline([('scaler', StandardScaler()), ('linreg', LinearRegression())])
pipe1.fit(X, y)
y_pred4 = pipe1.predict(X)
print(f"L2 loss for Linear Regression model is {sklearn.metrics.mean_squared_error(y, y_pred4):.3f}")
```

```python
pipe2 = sklearn.pipeline.Pipeline([('scaler', StandardScaler()), ('ridge', Ridge())])
pipe2.fit(X, y)
y_pred5 = pipe2.predict(X)
print(f"L2 loss for Ridge Regression model is {sklearn.metrics.mean_squared_error(y, y_pred5):.3f}")
```

```python
pipe3 = sklearn.pipeline.Pipeline([('scaler', StandardScaler()), ('lasso', Lasso())])
pipe3.fit(X, y)
y_pred6 = pipe3.predict(X)
print(f"L2 loss for Lasso Regression model is {sklearn.metrics.mean_squared_error(y, y_pred3):.3f}")
```

The L2 losses of the Ridge and Lasso models performed significantly better than in problem 6 due to the sensitivity of these models to scaling.
